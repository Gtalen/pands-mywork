```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Generating 1000 random numbers with normal distribution using numpy
values = np.random.normal(5, 2, 1000)    # mean = 5, SD = 2
print (values)

#plotting a histogram of the values

plt.hist(values, bins = 'auto', color = 'brown', edgecolor = 'black')
plt.title ('Normal distribution Histogram')
plt.xlabel ('values')
plt.ylabel('Frequency')
plt.text(120, 10, 'mean = 5, sd = 2')
plt.show()
```

```python
# h(x)=x3 in the range 0 to 10,
def h(x):
  return x ** 3
 
 #create x and y ranges

x = np.linspace (0, 10)
y = h(x)

#plotting the data using the function h(x)=x3 in the range 0 to 10

plt.plot (x,y)
plt.xlabel ('x in range 0 - 10')
plt.ylabel (' y = h(x)^3')
plt.legend()
plt.show ()

```
