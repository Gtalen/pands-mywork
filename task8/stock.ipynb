```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stocks = pd.read_csv('HistoricalPrices.csv')
stocks.head()
```

```python

#remnaming the heading using dictionary keys and values to strip the space at the beginning of each column heading
stocks = stocks.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
```

```python
print (stocks)
```

```python
#converting the date variable to dates using pandas datetime function and sort to sort the dates in ascending order
stocks['Date'] = pd.to_datetime(stocks['Date'])
stocks = stocks.sort_values(by = 'Date')
```

```python
#drawing line plots
plt.plot(stocks['Date'], stocks['Open'], label = 'Open')
plt.plot(stocks['Date'], stocks['Close'], label = 'Close')
plt.legend()
plt.show()
```
